**Fungsi Segmentasi gambar**

1.  Adaptive Thresholding ‚Üí buat kontras antara latar & sel

2.  Canny Edge Detection ‚Üí deteksi tepi sel darah

3.  Morphological Operations ‚Üí bersihin hasil segmentasi (noise removal)


**IMPORT LIBRARY**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

**PROSESS SEGMENTASI SEMUA DATA GAMBAR**

```python
def segment_blood_cells(image_path):
    # 1. Load gambar
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 2. Adaptive Thresholding (tingkatkan kontras)
    thresh = cv2.adaptiveThreshold(
        gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY_INV, 11, 2
    )

    # 3. Morphological operations (open untuk hilangkan noise kecil)
    kernel = np.ones((3, 3), np.uint8)
    morph = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

    # 4. Edge detection pakai Canny
    edges = cv2.Canny(morph, 100, 200)

    # 5. Optional: Contour detection (untuk bounding box/visualisasi)
    contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour_img = image_rgb.copy()
    cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 1)

    # 6. Tampilkan hasil
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 4, 1)
    plt.title("Original")
    plt.imshow(image_rgb)
    plt.axis("off")

    plt.subplot(1, 4, 2)
    plt.title("Threshold")
    plt.imshow(thresh, cmap='gray')
    plt.axis("off")

    plt.subplot(1, 4, 3)
    plt.title("Canny Edges")
    plt.imshow(edges, cmap='gray')
    plt.axis("off")

    plt.subplot(1, 4, 4)
    plt.title("Contours")
    plt.imshow(contour_img)
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```
