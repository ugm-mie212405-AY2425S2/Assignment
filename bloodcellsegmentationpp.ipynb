## Canny Edge Detection

```python
# Parameter threshold untuk Canny
low_threshold = 125
high_threshold = 300

# List untuk menyimpan hasil deteksi tepi
edge_eb = []
edge_bs = []
edge_nt = []
edge_es = []
edge_mn = []
edge_lp = []
edge_pt = []
edge_ig = []

# Proses setiap gambar grayscale
for gray_image in grayscale_eb:
    edges = cv2.Canny(gray_image, low_threshold, high_threshold)
    edge_eb.append(edges)
    
for gray_image in grayscale_bs:
    edges = cv2.Canny(gray_image, low_threshold, high_threshold)
    edge_bs.append(edges)
    
for gray_image in grayscale_nt:
    edges = cv2.Canny(gray_image, low_threshold, high_threshold)
    edge_nt.append(edges)
    
for gray_image in grayscale_es:
    edges = cv2.Canny(gray_image, low_threshold, high_threshold)
    edge_es.append(edges)
    
for gray_image in grayscale_mn:
    edges = cv2.Canny(gray_image, low_threshold, high_threshold)
    edge_mn.append(edges)
    
for gray_image in grayscale_lp:
    edges = cv2.Canny(gray_image, low_threshold, high_threshold)
    edge_lp.append(edges)
    
for gray_image in grayscale_pt:
    edges = cv2.Canny(gray_image, low_threshold, high_threshold)
    edge_pt.append(edges)
    
for gray_image in grayscale_ig:
    edges = cv2.Canny(gray_image, low_threshold, high_threshold)
    edge_ig.append(edges)

# Tampilkan 5 contoh hasil deteksi tepi
plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(edge_eb[i], cmap='gray')
    plt.title(f"Tepi {i+1}")
    plt.axis("off")

plt.show()
plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(edge_bs[i], cmap='gray')
    plt.title(f"Tepi {i+1}")
    plt.axis("off")

plt.show()
plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(edge_nt[i], cmap='gray')
    plt.title(f"Tepi {i+1}")
    plt.axis("off")

plt.show()
plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(edge_es[i], cmap='gray')
    plt.title(f"Tepi {i+1}")
    plt.axis("off")

plt.show()

plt.show()
plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(edge_mn[i], cmap='gray')
    plt.title(f"Tepi {i+1}")
    plt.axis("off")

plt.show()

plt.show()
plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(edge_lp[i], cmap='gray')
    plt.title(f"Tepi {i+1}")
    plt.axis("off")

plt.show()

plt.show()
plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(edge_pt[i], cmap='gray')
    plt.title(f"Tepi {i+1}")
    plt.axis("off")

plt.show()

plt.show()
plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(edge_ig[i], cmap='gray')
    plt.title(f"Tepi {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar dengan deteksi tepi: {len(edge_images)}")
```

## Erosi Dilasi 1

```python
# Kernel untuk operasi morfologi (3x3 berbentuk elips)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))

# List untuk menyimpan hasil dilasi dan erosi
processed_eb = []
processed_bs = []
processed_nt = []
processed_es = []
processed_mn = []
processed_lp = []
processed_pt = []
processed_ig = []

for edge in edge_eb:
    # Dilasi (memperbesar area tepi) sebanyak 2 kali
    dilated = cv2.dilate(edge, kernel, iterations=2)
    
    # Erosi (membersihkan noise) sebanyak 2 kali
    eroded = cv2.erode(dilated, kernel, iterations=2)
    
    # Simpan hasil akhir
    processed_eb.append(eroded)
    
for edge in edge_bs:
    # Dilasi (memperbesar area tepi) sebanyak 2 kali
    dilated = cv2.dilate(edge, kernel, iterations=2)
    
    # Erosi (membersihkan noise) sebanyak 2 kali
    eroded = cv2.erode(dilated, kernel, iterations=2)
    
    # Simpan hasil akhir
    processed_bs.append(eroded)

for edge in edge_nt:
    # Dilasi (memperbesar area tepi) sebanyak 2 kali
    dilated = cv2.dilate(edge, kernel, iterations=2)
    
    # Erosi (membersihkan noise) sebanyak 2 kali
    eroded = cv2.erode(dilated, kernel, iterations=2)
    
    # Simpan hasil akhir
    processed_nt.append(eroded)
    
for edge in edge_es:
    # Dilasi (memperbesar area tepi) sebanyak 2 kali
    dilated = cv2.dilate(edge, kernel, iterations=2)
    
    # Erosi (membersihkan noise) sebanyak 2 kali
    eroded = cv2.erode(dilated, kernel, iterations=2)
    
    # Simpan hasil akhir
    processed_es.append(eroded)
    
for edge in edge_mn:
    # Dilasi (memperbesar area tepi) sebanyak 2 kali
    dilated = cv2.dilate(edge, kernel, iterations=2)
    
    # Erosi (membersihkan noise) sebanyak 2 kali
    eroded = cv2.erode(dilated, kernel, iterations=2)
    
    # Simpan hasil akhir
    processed_mn.append(eroded)
    
for edge in edge_lp:
    # Dilasi (memperbesar area tepi) sebanyak 2 kali
    dilated = cv2.dilate(edge, kernel, iterations=2)
    
    # Erosi (membersihkan noise) sebanyak 2 kali
    eroded = cv2.erode(dilated, kernel, iterations=2)
    
    # Simpan hasil akhir
    processed_lp.append(eroded)
    
for edge in edge_pt:
    # Dilasi (memperbesar area tepi) sebanyak 2 kali
    dilated = cv2.dilate(edge, kernel, iterations=2)
    
    # Erosi (membersihkan noise) sebanyak 2 kali
    eroded = cv2.erode(dilated, kernel, iterations=2)
    
    # Simpan hasil akhir
    processed_pt.append(eroded)
    
for edge in edge_ig:
    # Dilasi (memperbesar area tepi) sebanyak 2 kali
    dilated = cv2.dilate(edge, kernel, iterations=2)
    
    # Erosi (membersihkan noise) sebanyak 2 kali
    eroded = cv2.erode(dilated, kernel, iterations=2)
    
    # Simpan hasil akhir
    processed_ig.append(eroded)
    

# Tampilkan 5 contoh hasil dilasi dan erosi
plt.figure(figsize=(12,6))
for i in range(min(5, len(processed_eb))):  # Pastikan tidak lebih dari jumlah gambar yang ada
    plt.subplot(1, 5, i+1)
    plt.imshow(processed_eb[i], cmap='gray')
    plt.title(f"Processed {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar setelah dilasi & erosi: {len(processed_eb)}")

# Tampilkan 5 contoh hasil dilasi dan erosi
plt.figure(figsize=(12,6))
for i in range(min(5, len(processed_bs))):  # Pastikan tidak lebih dari jumlah gambar yang ada
    plt.subplot(1, 5, i+1)
    plt.imshow(processed_bs[i], cmap='gray')
    plt.title(f"Processed {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar setelah dilasi & erosi: {len(processed_bs)}")

# Tampilkan 5 contoh hasil dilasi dan erosi
plt.figure(figsize=(12,6))
for i in range(min(5, len(processed_nt))):  # Pastikan tidak lebih dari jumlah gambar yang ada
    plt.subplot(1, 5, i+1)
    plt.imshow(processed_nt[i], cmap='gray')
    plt.title(f"Processed {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar setelah dilasi & erosi: {len(processed_nt)}")

# Tampilkan 5 contoh hasil dilasi dan erosi
plt.figure(figsize=(12,6))
for i in range(min(5, len(processed_es))):  # Pastikan tidak lebih dari jumlah gambar yang ada
    plt.subplot(1, 5, i+1)
    plt.imshow(processed_es[i], cmap='gray')
    plt.title(f"Processed {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar setelah dilasi & erosi: {len(processed_es)}")

# Tampilkan 5 contoh hasil dilasi dan erosi
plt.figure(figsize=(12,6))
for i in range(min(5, len(processed_mn))):  # Pastikan tidak lebih dari jumlah gambar yang ada
    plt.subplot(1, 5, i+1)
    plt.imshow(processed_mn[i], cmap='gray')
    plt.title(f"Processed {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar setelah dilasi & erosi: {len(processed_mn)}")

# Tampilkan 5 contoh hasil dilasi dan erosi
plt.figure(figsize=(12,6))
for i in range(min(5, len(processed_lp))):  # Pastikan tidak lebih dari jumlah gambar yang ada
    plt.subplot(1, 5, i+1)
    plt.imshow(processed_lp[i], cmap='gray')
    plt.title(f"Processed {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar setelah dilasi & erosi: {len(processed_lp)}")

# Tampilkan 5 contoh hasil dilasi dan erosi
plt.figure(figsize=(12,6))
for i in range(min(5, len(processed_pt))):  # Pastikan tidak lebih dari jumlah gambar yang ada
    plt.subplot(1, 5, i+1)
    plt.imshow(processed_pt[i], cmap='gray')
    plt.title(f"Processed {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar setelah dilasi & erosi: {len(processed_pt)}")

# Tampilkan 5 contoh hasil dilasi dan erosi
plt.figure(figsize=(12,6))
for i in range(min(5, len(processed_ig))):  # Pastikan tidak lebih dari jumlah gambar yang ada
    plt.subplot(1, 5, i+1)
    plt.imshow(processed_ig[i], cmap='gray')
    plt.title(f"Processed {i+1}")
    plt.axis("off")

plt.show()

print(f"Total gambar setelah dilasi & erosi: {len(processed_ig)}")

```

## Filling Holes

```python
# List untuk menyimpan hasil akhir setelah pembersihan objek
cleaned_eb = []
cleaned_bs = []
cleaned_nt = []
cleaned_es = []
cleaned_mn = []
cleaned_lp = []
cleaned_pt = []
cleaned_ig = []

def clean_images(processed_images):
    cleaned = []
    for img in processed_images:
        # Temukan kontur objek
        contours, _ = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        # Buat masker kosong dengan ukuran gambar
        mask = np.zeros_like(img)

        # Filter objek berdasarkan area (misal, hanya objek lebih besar dari 500 px yang diambil)
        for contour in contours:
            if cv2.contourArea(contour) > 500:
                cv2.drawContours(mask, [contour], -1, (255), thickness=cv2.FILLED)

        # Simpan hanya jika ada piksel putih (tidak sepenuhnya hitam)
        if np.any(mask > 10):
            cleaned.append(mask)
    
    return cleaned

# Terapkan fungsi ke semua kategori gambar
cleaned_eb = clean_images(final_processed_eb)
cleaned_bs = clean_images(final_processed_bs)
cleaned_nt = clean_images(final_processed_nt)
cleaned_es = clean_images(final_processed_es)
cleaned_mn = clean_images(final_processed_mn)
cleaned_lp = clean_images(final_processed_lp)
cleaned_pt = clean_images(final_processed_pt)
cleaned_ig = clean_images(final_processed_ig)

# Cek jumlah gambar setelah pembersihan
print(f"Total gambar setelah pembersihan objek EB: {len(cleaned_eb)}")
print(f"Total gambar setelah pembersihan objek BS: {len(cleaned_bs)}")
print(f"Total gambar setelah pembersihan objek NT: {len(cleaned_nt)}")
print(f"Total gambar setelah pembersihan objek ES: {len(cleaned_es)}")
print(f"Total gambar setelah pembersihan objek MN: {len(cleaned_mn)}")
print(f"Total gambar setelah pembersihan objek LP: {len(cleaned_lp)}")
print(f"Total gambar setelah pembersihan objek PT: {len(cleaned_pt)}")
print(f"Total gambar setelah pembersihan objek IG: {len(cleaned_ig)}")

# Tampilkan contoh hasil setelah pembersihan
def show_images(cleaned_images, title):
    plt.figure(figsize=(12,6))
    for i in range(min(10, len(cleaned_images))):
        plt.subplot(1, 10, i+1)
        plt.imshow(cleaned_images[i], cmap='gray')
        plt.title(f"{title} {i+1}")
        plt.axis("off")
    plt.show()

show_images(cleaned_eb, "EB")
show_images(cleaned_bs, "BS")
show_images(cleaned_nt, "NT")
show_images(cleaned_es, "ES")
show_images(cleaned_mn, "MN")
show_images(cleaned_lp, "LP")
show_images(cleaned_pt, "PT")
show_images(cleaned_ig, "IG")
```
