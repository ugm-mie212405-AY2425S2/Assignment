```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Path folder tempat citra disimpan
input_folder = "bloodCell/"  # Ganti dengan nama folder yang sesuai
output_folder = "hasil_segmentasi/"  # Folder untuk menyimpan hasil
```

```python
# Pastikan folder output ada, kalau tidak buat
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Ambil semua file dalam folder
image_files = [f for f in os.listdir(input_folder) if f.endswith((".jpg", ".png", ".jpeg"))]
```

```python
# Proses setiap gambar dalam folder
for img_name in image_files:
    # Baca gambar
    img_path = os.path.join(input_folder, img_name)
    image = cv2.imread(img_path)
    
    if image is None:
        print(f"Gagal membaca {img_name}, skip...")
        continue
    
    # Konversi ke grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Otsu's Thresholding
    _, thresholded = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Simpan hasil segmentasi
    output_path = os.path.join(output_folder, f"otsu_{img_name}")
    cv2.imwrite(output_path, thresholded)

    # Tampilkan hasil untuk verifikasi
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(gray, cmap="gray")
    plt.title(f"Grayscale - {img_name}")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(thresholded, cmap="gray")
    plt.title(f"Otsu - {img_name}")
    plt.axis("off")

    plt.show()

print("Segmentasi selesai. Hasil disimpan di folder:", output_folder)
```
